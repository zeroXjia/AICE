**`집합(set)`**

**`딕셔너리`**
- {key1: value1, key2: value2, key3: value3}의 형태를 가짐
- 순서의 의미가 없으며, key를 통해 value를 확인
- 인덱싱x, 슬라이싱x
- <span style="color:palevioletred">**in**</span> 연산자를 사용해 해당 <span style="color:palevioletred">**key**</span>가 있는지 확인 가능
- 없는 key로 변경하면 key:value 형태로 추가됨

```python
# in 

# 이름 확인
member = {'이름': '홍길동', '취미': ['독서', '여행', '걷기']}
print(member['이름'])

# 데이터 존재 여부 확인
print('이름' in member)     # True
print('홍길동' in member)   # False
```

**배열을 만드는 함수들**

```python
# 숫자간 변환 astype
# 배열 -> 리스트 변환   .tolist()사용 권고

# 배열 만들기
x = np.array([1, 2])
print(type(x))    # numpy.ndarray

# 배열 --> 리스트
y = x.tolist()
print(y)          # [1, 2]
print(type(y))    # list
```

# 4. pandas

```python
# 라이브러리 불러오기
import pandas as pd
```

**시리즈 만들기**
- pd.Series()함수를 사용해 다른 자료형 데이터로 시리즈를 만듦

```python
# 리스트 --> 시리즈
list_stock = [92300, 94300, 92100, 92400, 92600]
stock = pd.Series(list_stock)
stock
```

```python
# 딕셔너리 --> 시리즈
dict_stock = {'2019-02-15': 92300
             ,'2019-02-16': 94300
             ,'2019-02-17': 92100
             ,'2019-02-18': 92400
             ,'2019-02-19': 92600}
stock = pd.Series(dict_stock)
stock
```

**loc vs iloc**
- 특정 <span style="color:palevioletred">**위치**</span>의 행 조회할 때 => <span style="color:palevioletred">**iloc**</span> 사용
- 특정 <span style="color:palevioletred">**인덱스 값**</span>을 갖는 행 조회 => <span style="color:palevioletred">**loc**</span> 사용
